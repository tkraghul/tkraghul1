# Task 1: Fitting to a Periodic Function and Plotting the Fit

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
months = np.arange(12)
max_temps = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
min_temps = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
```

```python
def periodic_func(x, a, b, c):
    return a + b * np.sin((x + c) * 2 * np.pi / x.max())
```

```python
popt_max, pcov_max = curve_fit(periodic_func, months, max_temps)
popt_min, pcov_min = curve_fit(periodic_func, months, min_temps)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(months, max_temps, 'ro', label='Max Temp')
plt.plot(months, min_temps, 'bo', label='Min Temp')
plt.plot(months, periodic_func(months, *popt_max), 'r--', label='Max Temp Fit')
plt.plot(months, periodic_func(months, *popt_min), 'b--', label='Min Temp Fit')
plt.xlabel('Months')
plt.ylabel('Temperature (°C)')
plt.title('Fit to Periodic Function for Max and Min Temperatures')
plt.legend()
plt.show()
```

# Task 2: Creating Pie Chart and Scatterplot

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)
```

```python
gender_counts = titanic['sex'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Male/Female Proportion')
plt.show()
```

```python
titanic_cleaned = titanic.dropna(subset=['sex'])
```

```python
colors = {'male': 'blue', 'female': 'red'}
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(titanic_cleaned['age'], titanic_cleaned['fare'], c=titanic_cleaned['sex'].map(colors), alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatterplot of Fare vs Age (Colored by Gender)')
plt.show()
```
